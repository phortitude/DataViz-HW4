### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
# create data frame to store purchase data object
purchase_data_df = pd.DataFrame(purchase_data)
# preview purchase_data data frame
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
# perform nunique function on the "SN" column of purchase_data and assign values to variable
count_total_players = purchase_data["SN"].nunique()
# pass values from variable containing unique players to a data frame and put data frame in a variable
players = pd.DataFrame({"Total Players": [count_total_players]})

total_players = players["Total Players"][0]

total_players

```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# find unique items from Item ID and pass values to variable
unique_items = purchase_data_df["Item ID"].nunique()
# calculate the average price
average_price = purchase_data_df["Price"].mean()
# calculate the total number of purchases
number_of_purchases = purchase_data_df["Price"].count()
# calculate total revenue
total_revenue = purchase_data_df["Price"].sum()

# unique_items
# average_price
# number_of_purchases
# total_revenue
# create purchasing analysis data frame to display results
sales_analysis_df = pd.DataFrame({"Unique Items": [unique_items], "Average Price": [average_price], "Number Purchases": [number_of_purchases], "Total Revenue": [total_revenue]})

sales_analysis_df

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# group by gender and pass values to variable
gender_groups = purchase_data_df.groupby(["Gender"])
# calculate gender count
gender_count = gender_groups["SN"].nunique()
# calculate gender percentages
per_gender = ((gender_count / count_total_players) *100)
# create data frame that displays total count and percentage of players
gender_df = pd.DataFrame({"Total Count": gender_count, 
                          "Percentage of Players": per_gender})
gender_df = gender_df.round(2)
gender_df

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# calculate purchase count
purch_count = gender_groups["Item ID"].count()
# calculate average purchase price
av_purch_price = purchase_data_df.groupby("Gender")["Price"].mean()
# calcuate total purchase value
tot_purch_value = purchase_data_df.groupby("Gender")["Price"].sum()
# formate total purchase value
formatted_purch_value = tot_purch_value.map('${:,.2f}'.format)
# calculate average purchase total per person
av_tot_purch_person = tot_purch_value/gender_count
# create data frame to display results
form_purch_value = formatted_purch_value
purchase_analysis_df = pd.DataFrame({"Purchase Count": purch_count, 
                          "Average Purchase Price": av_purch_price,
                           "Total Purchase Value": form_purch_value,
                           "Average Total Purchase per Person": av_tot_purch_person})
purchase_analysis_df = purchase_analysis_df.round(2)
purchase_analysis_df.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# get max age
max_age = purchase_data_df["Age"].max()
min_age = purchase_data_df["Age"].min()

# create bins for ages
bins = [0, 10, 15, 20, 25, 30, 35, 40, 46]
# set group name
group_name = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
# create the age data categories using pd.cut and create new column in purchase data
purchase_data_df["Age Category"] = pd.cut(purchase_data_df["Age"], bins, labels= group_name)
purchase_data_df.head()
```

```python
# calculate numbers and percentages by age group
age_group_totals = purchase_data_df['Age Category'].value_counts()
# create a df to hold the object's values
age_group_totals_df = pd.DataFrame(age_group_totals)
# create column headers
age_group_totals_df.columns=["Age Category Total"]
age_group_totals_df.sort_index()
```

```python
percent_players = ((age_group_totals / count_total_players) * 100)
# create a df to hold the object's values
percent_players_df = pd.DataFrame(percent_players)
# create column headers
percent_players_df.columns=["Percentage of Players"]
# format the numbers
per_players = percent_players_df.round(2)
per_players.sort_index()
```

```python
# merge two dataframes with outer join
# DataFrame.merge(right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), copy=True, indicator=False, validate=None)

summary_results_df = pd.merge(age_group_totals_df, per_players, left_index=True, right_index=True)
summary_results_df.sort_index()

#print(type(age_group_totals_df))
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# purchase_data_df["Age Category"] = pd.cut(purchase_data_df["Age"], bins, labels= group_name)
purchase_average = purchase_data_df.groupby(["Age Category"]).mean()["Price"]

purchase_total = purchase_data_df.groupby(["Age Category"]).sum()["Price"]

purchase_count = purchase_data_df.groupby(["Age Category"]).count()["Price"]
purchase_count

summary_purch_analysis_df = pd.DataFrame({'purchase_average': purchase_average, 'purchase_count': purchase_count, 
                                          'purchase_total': purchase_total})

summary_purch_analysis_df.sort_index()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# create group by dataframe
total_price_by_sn = purchase_data_df.groupby("SN").sum()["Price"]
av_purchase_by_sn = purchase_data_df.groupby("SN").mean()["Price"]
purchase_count_by_sn = purchase_data_df.groupby("SN").count()["Price"]

top_spenders_df = pd.DataFrame({'Purchase Count': purchase_count_by_sn, 
                               'Average Purchase Price': av_purchase_by_sn, 
                               'Total Purchase Value': total_price_by_sn}).reset_index()

top_spenders_sorted = top_spenders_df.sort_values(['Total Purchase Value'], ascending=False) 

top_spenders_sorted.head(20)

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_data_df.head()

purchase_count = purchase_data_df.groupby(["Item ID", "Item Name"]).count()["Price"]
item_price = purchase_data_df.groupby(["Item ID", "Item Name"]).mean()["Price"]
total_purchase_value = purchase_data_df.groupby(["Item ID", "Item Name"]).sum()["Price"]

popular_items_df = pd.DataFrame({'Purchase Count': purchase_count,
                                     'Item Price': item_price, 
                                     'Total Purchase Value': total_purchase_value}).reset_index()

popular_items_df.head(20)
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
most_profitable = popular_items_df.sort_values("Total Purchase Value", ascending=False).reset_index()

most_profitable.head(20)
```
